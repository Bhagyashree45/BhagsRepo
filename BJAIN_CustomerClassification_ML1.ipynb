```python
# Convert customer data into Dataframe
customer_data = customer_data.drop(['REF_NO','self_employed','self_employed_partner','post_code','post_area'], axis=1)
df = pd.DataFrame(customer_data) 
df.head()
df.shape
```

```python
# Creating dummy variables
df = df.drop('region', axis=1) 
df = pd.get_dummies(df) 
df.head()
df.shape
```

```python
df.shape
```

```python
# Spliting data in train test for modeling 
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(df, df.Revenue_Grid, test_size=0.25, random_state = 0)
x_train = x_train.drop('Revenue_Grid', 1) 
x_test = x_test.drop('Revenue_Grid', 1)
```

```python
# Declaring LR
from sklearn.linear_model import LogisticRegression
# all parameters not specified are set to their defaults
logisticRegr = LogisticRegression()
```

```python
# Running model to fit training dataset
logisticRegr.fit(x_train, y_train)
```

```python
# Predictions using test dataset
predictions = logisticRegr.predict(x_test)
print(predictions)
```

```python
# Use score method to get accuracy of model
score = logisticRegr.score(x_test, y_test)
print('Accuracy of the Model: (in percentage)')
print(score*100)
```

```python
# Confusion Matrix
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
# f1 score on trained model
f1_score(y_test,predictions, average=None)
```
