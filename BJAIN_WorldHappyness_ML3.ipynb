```python
# Convert  data into Dataframe
df = pd.DataFrame(customer_data)
df = df.drop(['Country name','ID'], axis = 1)
df.head()
```

```python
# Creating dummy variables
df = pd.get_dummies(df)
df['Social support'] = df['Social support'].fillna(0)
df['Freedom to make life choices'] = df['Freedom to make life choices'].fillna(0)
df['Generosity'] = df['Generosity'].fillna(0)
df['Perceptions of corruption'] = df['Perceptions of corruption'].fillna(0)
df['Positive affect'] = df['Positive affect'].fillna(0)
df['Negative affect'] = df['Negative affect'].fillna(0)
df['Log GDP per capita'] = df['Log GDP per capita'].fillna(0)
df['year'] = df['year'].fillna(0)
df['Healthy life expectancy at birth'] = df['Healthy life expectancy at birth'].fillna(0)
df.head()
```

```python
# Spliting data in train test for modeling
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, df['Life Ladder'], test_size=0.25, random_state=0)
x_train = x_train.drop('Life Ladder', 1)
x_test = x_test.drop('Life Ladder', 1)
```

```python
# Declaring LR
from sklearn.linear_model import LinearRegression
# all parameters not specified are set to their defaults
linear_regression = LinearRegression()
```

```python
# Running model to fit training dataset
reg = LinearRegression().fit(x_train, y_train)
reg.score(x_train, y_train)
```

```python
# make predictions
x_test_result = reg.predict(x_test)
# evaluate predictions
mae = mean_absolute_error(y_test, x_test_result)
print('MAE: %.3f' % mae)
```
